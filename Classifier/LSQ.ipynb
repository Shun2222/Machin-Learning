```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
import seaborn as sns; sns.set()

sns.set_context('talk')
```

## データ生成

```python
# 正規分布
def GaussDensity(x, mu=0.0, sgm=1.0):
    return 1/np.sqrt(2 * np.pi)/sgm * np.exp(-0.5 * ((x - mu)/sgm)**2 ) 
    

np.random.seed(12345)

N1 = 20
N2 = 40

# とりあえず、等分散
mu1 = 0.0
mu2 = 1.0
sgm1 = sgm2 = 0.25

# 等分散より，真の識別境界は mu1 と mu2 の中点
x1 = sgm1 * np.random.randn(N1) + mu1
x2 = sgm2 * np.random.randn(N2) + mu2

# データの表示
plt.plot(x1, np.zeros(N1), 'ro', alpha=0.8)
plt.plot(x2, np.zeros(N2), 'bo', alpha=0.8)

xx = np.linspace(-1, 2, 256)
yy1 = GaussDensity(xx, mu=mu1, sgm=sgm1)
yy2 = GaussDensity(xx, mu=mu2, sgm=sgm2)
plt.plot(xx, yy1, 'r-', linewidth=3)
plt.plot(xx, yy2, 'b-', linewidth=3)
```

## label付け

```python
# label付け
y1 = -np.ones(N1)
y2 = np.ones(N2)

# 結合
x = np.hstack((x1, x2))
y = np.hstack((y1, y2))

# 表示
plt.plot(x1, y1, 'ro')
plt.plot(x2, y2, 'bo')
```

```python

def Solver1(x, y):
    # 中心化．
    Ex = np.mean(x)
    Ey = np.mean(y)
    xnew = x - Ex
    ynew = y - Ey
    
    # 線形として傾き、切片を求める
    Exx = np.mean(xnew**2)
    Exy = np.mean(xnew * ynew)
    w0 = 0
    w1 = Exy / Exx
    
    # 元の座標系に戻す
    w0 = w0 + Ey - w1 * Ex

    # 分離境界は y = w_0 + w_1 x = 0 となる x 
    xbound = -w0 / w1

    return w0, w1, xbound


w0, w1, xbound = Solver1(x, y)

# グラフ
yy = w0 + w1 * xx

plt.plot(x[y==+1], y[y==+1], 'ro', alpha=0.8)
plt.plot(x[y==-1], y[y==-1], 'bo', alpha=0.8)

plt.plot(xx, yy, 'g-', linewidth=3) # 回帰直線

plt.plot(xbound, 0, 'go') 
plt.plot(0.5, 0, 'mo') # 真の境界

plt.xlim(-1, 2)
plt.ylim(-2, 2)
print('w0: {}, w1: {}, xbound: {}'.format(w0, w1, xbound))
```
