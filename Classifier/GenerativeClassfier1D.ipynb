# １次元ガウス分布 Posterior
確率的生成モデルを用いてパターン認識を行う

```python
import numpy as np
import matplotlib.pylab as plt
import matplotlib.gridspec as gridspec
import seaborn as sns; sns.set()
```

## データ生成

```python
# 各クラスの生成確率
q = 0.3
p1 = q
p2 = 1-q
```

```python
# 各正規分布の中心
mu1 = 1
mu2 = -1

# 分散は共通にする
sgm = 0.8

# 1次元のガウス分布の定義
def gauss1d(x, mu, sig):
    return np.sqrt(1./(2.*np.pi*sig**2)) * np.exp( - 0.5 * ( (x-mu)/sig )**2 )
```

```python
# データ数
N = 100

# 2項分布からラベルデータを作成
xlabel = np.random.binomial(1, p1, N)


# 生成したラベル xlabel に基づき，各クラスのデータを生成
x = np.zeros(N)
x1num = xlabel[xlabel==1].sum()
x2num = N - x1num

x[xlabel==1] = mu1 + sgm * np.random.randn(x1num)
x[xlabel!=1] = mu2 + sgm * np.random.randn(x2num)
```

```python
xrng = (-4, 4)

gs = gridspec.GridSpec(2, 1, height_ratios=[4, 1])
ax0 = plt.subplot(gs[0])

plt.title('Conditional Densities')
xx = np.linspace(xrng[0], xrng[1], 128)
plt.plot(xx, gauss1d(xx, mu1, sgm), 'r-', alpha=0.8)
plt.plot(xx, gauss1d(xx, mu2, sgm), 'b-', alpha=0.8)
plt.xlim(xrng)

ax1 = plt.subplot(gs[1])
plt.axhline(y=0, color='black')
plt.plot(x[xlabel==1], np.zeros(x1num), 'ro', alpha=0.8)
plt.plot(x[xlabel!=1], np.zeros(x2num), 'bx', alpha=0.8)
plt.xlim(xrng)
```

```python
def posterior(x, mu1, mu2, s1, s2, p1, p2):
    j1 = gauss1d(x, mu1, s1) * p1
    j2 = gauss1d(x, mu2, s2) * p2
    return j1 / (j1 + j2)


plt.subplot(2, 1, 1)
plt.plot(xx, posterior(xx, mu1, mu2, sgm, sgm, p1, p2), 'r-')
plt.plot(xx, posterior(xx, mu2, mu1, sgm, sgm, p2, p1), 'b-')
plt.xlim(xrng)

plt.subplot(2, 1, 2)
# ヒストグラムを下に描画してみる
nbins = np.linspace(xrng[0], xrng[1], 16)
plt.hist(x[xlabel==1], bins=nbins, density=False, color='red', alpha=0.6)
plt.hist(x[xlabel!=1], bins=nbins, density=False, color='blue', alpha=0.6)
```
