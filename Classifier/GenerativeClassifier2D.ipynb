# ２次元版 ガウス生成モデル


```python
import numpy as np
import scipy as sp
import matplotlib.pylab as plt
import matplotlib.gridspec as gridspec
import matplotlib.cm as cm
import seaborn as sns; sns.set()
#sns.set_context('talk')
```

## データ生成

```python
def gfunc2d(x, y, mu, cov):
    xz = x - mu[0]
    yz = y - mu[1]
    dlt = cov[0,0] * xz**2 + 2 * cov[0,1] * xz*yz + cov[1,1]* yz**2
    det = cov[0,0] * cov[1,1] - cov[0,1]**2
    return np.sqrt(1. / (2.*np.pi*det)) * np.exp(-0.5 * dlt)

def posterior(x, y, m1, m2, s1, s2, p1, p2):
    o1 = gfunc2d(x, y, m1, s1) * p1
    o2 = gfunc2d(x, y, m2, s2) * p2
    return o1/(o1+o2)
```

```python
N1 = 30 * 2
N2 = 70 * 2
p1 = np.double(N1) / (N1 + N2)
p2 = np.double(N2) / (N1 + N2)

# メッシュグリッドを作って等高線をかけるようにする
xymin = -7
xymax = +7
t = np.linspace(xymin, xymax, 256)
xx, yy = np.meshgrid(t, t)
```

```python
# ガウス分布の設定
th1 = np.pi/3.
mu1 = np.array([[1.], [1.]]) * 0.7
mu2 = np.array([[-1.], [-1.]]) * 0.7
s1 = 2.0
s2 = 0.2
ss = np.array([[1./s1, 0.], [0, 1./s2]])
c1 = np.cos(th1)
s1 = np.sin(th1)
rot = np.array([[c1, -s1], [s1, c1]])
cv = rot.T @ ss @ rot
cvInv = np.linalg.inv(cv)

zz1 = gfunc2d(xx, yy, mu1, cv)
zz2 = gfunc2d(xx, yy, mu2, cv)
```

```python
# ２次元ガウス分布の表示
plt.figure(figsize=(8,8))
plt.contour(xx, yy, zz1, 10, cmap='YlOrRd')
plt.contour(xx, yy, zz2, 10, cmap='YlGnBu')

Xdat1 = np.random.multivariate_normal(mu1.flatten(), cvInv, size=N1)
plt.plot(Xdat1[:,0], Xdat1[:,1], 'ro', alpha=0.6)

Xdat2 = np.random.multivariate_normal(mu2.flatten(), cvInv, size=N2)
plt.plot( Xdat2[:,0], Xdat2[:,1], 'bo', alpha=0.6)

plt.xlim((xymin, xymax))
plt.ylim((xymin, xymax))
```

## posterorを適応

```python
uu1 = posterior( xx, yy, mu1, mu2, cv, cv, p1, p2 )

plt.figure(figsize=(8, 8))
plt.contourf(xx, yy, uu1, cmap='jet', alpha=0.2)

plt.title('Posterior of C1')
plt.plot(Xdat1[:,0], Xdat1[:,1], 'ro', alpha=0.6)
plt.plot(Xdat2[:,0], Xdat2[:,1], 'bo', alpha=0.6)

plt.xlim((xymin, xymax))
plt.ylim((xymin, xymax))
```

## 分散構造を変える

```python
th1 = np.pi/3.
mu1 = np.array([[1.], [1.]]) * 1.2
mu2 = np.array([[-1.], [-1.]]) * 1.2
s1 = 2.0
s2 = 1.0
ss = np.array([[1./s1, 0.], [0, 1./s2]])
c1 = np.cos(th1)
s1 = np.sin(th1)
rot = np.array([[c1, -s1], [s1, c1]])
cv1 = rot.T @ ss @ rot
cv1Inv = np.linalg.inv(cv1)
cv2 = rot @ ss @ rot.T
cv2Inv = np.linalg.inv(cv2)

zz1 = gfunc2d(xx, yy, mu1, cv1)
zz2 = gfunc2d(xx, yy, mu2, cv2)
```

```python
plt.figure(figsize=(8,8))
plt.contour(xx, yy, zz1, 10, cmap='YlOrRd')
plt.contour(xx, yy, zz2, 10, cmap='YlGnBu')

Xdat1 = np.random.multivariate_normal(mu1.flatten(), cv1Inv, size=N1)
plt.plot(Xdat1[:,0], Xdat1[:,1], 'ro', alpha=0.6)

Xdat2 = np.random.multivariate_normal(mu2.flatten(), cv2Inv, size=N2)
plt.plot( Xdat2[:,0], Xdat2[:,1], 'bo', alpha=0.6)

plt.xlim((xymin, xymax))
plt.ylim((xymin, xymax))
```

```python
uu1 = posterior(xx, yy, mu1, mu2, cv1, cv2, p1, p2)

plt.figure(figsize=(8, 8))
plt.contourf(xx, yy, uu1, cmap='jet', alpha=0.2)
# plt.contour( xx, yy, uu1, 3, colors='r', origin='lower', extent=(-4.,4.,-4.,4.)  )
plt.title( 'Posterior of C1' )
plt.plot( Xdat1[:,0], Xdat1[:,1], 'ro' )
plt.plot( Xdat2[:,0], Xdat2[:,1], 'bo' )

plt.xlim((xymin, xymax))
plt.ylim((xymin, xymax))
```
