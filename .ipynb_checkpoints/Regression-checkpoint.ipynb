## 線形回帰モデルと最小二乗法

```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns; sns.set()
```

```python
# 真のモデル
def LinearModel(a, b, x, N=30):
    return a * x + b

# 観測データ
def ObsData(yrue, sgm=0.1, N=30):
    return ytrue  + sgm * np.random.randn(N)  # 真値にノイズを乗せる

# 最小二乗法のコスト関数
def E(a, b, xobs, yobs):
    Ex = np.mean(xobs)
    Ey = np.mean(yobs)
    Exx = np.mean(xobs**2)
    Eyy = np.mean(yobs**2)
    Exy = np.mean(xobs * yobs)
    
    return 0.5 * (Eyy - 2 * a * Exy - 2 * b * Ey + a**2 * Exx + 2 * a * b * Ex + b**2)

# データ生成
atrue = 1.2
btrue = 0.5
N = 40
x = np.random.rand(N)
ytrue = LinearModel(atrue, btrue, x, N)
y = ObsData(ytrue, 0.2, N)
print(np.size(y))
```

```python
# 散布図を表示
plt.plot(x, y, 'bo', label='data')


# 真値も描画しておく
xx = np.linspace(0, 1, 128)
yy = LinearModel(atrue, btrue, xx, N)

plt.plot(xx, yy, 'r-', label='true')
plt.xlim(0, 1)
plt.ylim(0.4, 1.8)
plt.legend()
```

```python
# 前処理 中心化
Ex = np.mean(x)
Ey = np.mean(y)
x = x - Ex
y = y - Ey
```

```python
# 観測データ 散布図
plt.plot(x, y, 'bo', label='data')


# 真値の中心化に合わせる
xx = x
yy = ytrue - Ey

plt.plot(xx, yy, 'r-', label='true')
plt.xlim(-0.5, 0.5)
plt.ylim(-0.7, 0.7)
plt.legend()
```
