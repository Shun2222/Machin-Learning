# Bayes推定の可視化
ベイズ推定では事前分布としてあらかじめ知っている知識を取り込められる。そのため、データが少ない場合に有用。<br>
事前分布<br>
$$
p(w_0, w_1) = \sqrt{\frac{\alpha}{2\pi}} \exp\left(-\frac{\alpha}{2} (w_0^2 + w_1^2)\right)
$$

```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns; sns.set()
from sklearn.preprocessing import PolynomialFeatures

# sns.set_context('talk')

# pcolormesh と pcolor の deprecated 問題を封じておく
plt.rcParams['axes.grid'] = False

plt.rcParams['font.size'] = 20
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
plt.rcParams['image.cmap'] = 'plasma'
# sns.set_context('talk')
```

## データ生成

```python
def model_func(x, w):
    phi = design_mat(x, w.shape[0])
    return(phi @ w)

def design_mat(x, Mdim):
    Ndim = x.shape[0]
    poly = PolynomialFeatures(Mdim-1)   # 計画行列
    return poly.fit_transform(x.reshape(Ndim, 1))
```

```python
N = 10

# モデル真値
a_true = 0.9
b_true = 0.2
sigma_true = 0.1
beta_true = 1/sigma_true**2

# 真値わからないのでそれっぽいの
alpha_true = 1.0

# alpha, beta　推定しない
beta = beta_true
alpha = alpha_true

w_true = np.array([b_true, a_true])

# 乱数シード固定
np.random.seed(54321)

# 真値と観測データ
x = np.random.rand(N)
y_true = model_func(x, w_true)
epsilon = sigma_true * np.random.randn(N)
y = y_true + epsilon
```

```python
# データの表示
xx = np.linspace(-0.1, 1.2)
yy = model_func(xx, w_true)
plt.figure()
plt.plot(xx, yy)

for n in range(N):
    plt.plot(x[n], y[n], 'ro', alpha=0.5)
    plt.annotate(n, xy=(x[n], y[n]))

plt.grid()
```

## 推定と表示


```python
# prior
def logridge_prior(w, alpha):
    return -alpha/2 * np.sum(w**2)


# 対数尤度関数
def log_likelihood(w, beta, x, y):
    Ndim = x.shape[0]
    const = Ndim/2 * np.log(beta/(2*np.pi))
    loss = np.sum((y - model_func(x, w))**2) 
    return -beta / 2 * loss + const
```

```python
# w0, w1 のパラメータ空間にメッシュを作成
delta = 128   
w0lim = (-1, 1)
w1lim = (0, 2)  

w0shp = np.linspace(w0lim[0], w0lim[1], delta)
w1shp = np.linspace(w1lim[0], w1lim[1], delta)
w1_p, w0_p = np.meshgrid(w1shp, w0shp)

w = np.array([w0_p.ravel(), w1_p.ravel()])

Wary = np.array(w)
M = Wary.shape[1]

# 事前分布
plist = [logridge_prior(Wary[:, m], alpha) for m in range(M)]
zpri = np.array(plist).reshape(delta, delta)


# 尤度関数
llist = [log_likelihood(Wary[:, m], beta_true, x, y) for m in range(M)]

# 等高線
zlike = np.array(llist).reshape(delta, delta)

plt.figure(figsize=(10, 8))
plt.ylim(w1lim[0], w1lim[1])
plt.xlim(w0lim[0], w0lim[1])

plt.contourf(w0_p, w1_p, np.exp(zlike + zpri),  alpha=0.6)  #cm.gist_heatにするとかっこいい
plt.colorbar()
plt.grid()

plt.title('Posterior Density')
plt.scatter(w_true[0], w_true[1], s=100, c='g')
```

## 事前分布の影響を見てみる
最尤法は1点だけでは確率が直線で高かったが、ベイズ推定はpriorのおかげで、ある程度確率がpriorに引っ張られている。そのため、1点でも最尤法よりも狭い範囲で推定している。

```python

# w0, w1 のパラメータ空間にメッシュを切る ここは前回のを流用
delta = 128   # メッシュの細かさ
w0lim = (-3, 3) # 真値が 0.2
w1lim = (-1, 5)  # 真値が 0.9

w0shp = np.linspace(w0lim[0], w0lim[1], delta)
w1shp = np.linspace(w1lim[0], w1lim[1], delta)
w1_p, w0_p = np.meshgrid(w1shp, w0shp)

w = np.array([w0_p.ravel(), w1_p.ravel()])
Wary = np.array(w)
M = Wary.shape[1]

# 事前分布
plist = [logridge_prior(Wary[:, m], alpha) for m in range(M)]
zpri = np.array(plist).reshape(delta, delta)


# zcum = np.zeros((delta, delta))
zcum = zpri

plt.figure(figsize=(10, 8))
plt.ylim(w1lim[0], w1lim[1])
plt.xlim(w0lim[0], w0lim[1])

plt.contourf(w0_p, w1_p, np.exp(zcum),  alpha=0.6)  
plt.colorbar()
plt.grid()

plt.scatter(w_true[0], w_true[1], c='g', s=100)
plt.title('Prior Density')

for n in range(1):
    M = w.shape[1]
    xx = x[n:n+1]
    yy = y[n:n+1]
    llist = [log_likelihood(Wary[:, m], beta_true, xx, yy) for m in range(M)]

    z = np.array(llist).reshape(delta, delta)
    zcum = zcum + z

    plt.figure(figsize=(10, 8))

    plt.ylim(w1lim[0], w1lim[1])
    plt.xlim(w0lim[0], w0lim[1])

    # 尤度関数の等高線の描画
    plt.contourf(w0_p, w1_p, np.exp(z),  alpha=0.6)  #cm.gist_heatにするとかっこいい
    plt.colorbar()
    plt.grid()

    # 真値の描画
    plt.scatter(w_true[0], w_true[1], s=100, c='g')
    plt.title('Likelihood {:02d}'.format(n))

    
    plt.figure(figsize=(10, 8))
    plt.ylim(w1lim[0], w1lim[1])
    plt.xlim(w0lim[0], w0lim[1])

    # 事後密度の等高線の描画
    plt.contourf(w0_p, w1_p, np.exp(zcum),  alpha=0.6)  #cm.gist_heatにするとかっこいい
    plt.colorbar()
    plt.grid()

    # 真値の描画
    plt.scatter(w_true[0], w_true[1], s=100, c='g')
    plt.title('Cum. Posterior upto {}'.format(n))
```
