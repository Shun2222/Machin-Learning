## ベイズ推定で回帰モデルの推定

```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns; sns.set()
from sklearn.preprocessing import PolynomialFeatures

# sns.set_context('talk')

# pcolormesh と pcolor の deprecated 問題を封じておく
plt.rcParams['axes.grid'] = False

plt.rcParams['font.size'] = 20
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
plt.rcParams['image.cmap'] = 'plasma'
```

多項式モデル用に画行列 $\Phi$ （今回の計画行列はデータ点すべてに対する冪を並べたもの）

$
\Phi = \begin{pmatrix}
1     & 1        & \ldots & 1 \\
x_1   & x_2      & \ldots & x_N \\
x_1^2 & x_2^2    & \ldots & x_N^2 \\
\vdots & \vdots  & \ddots & \vdots \\
x_1^M  & x_2^M   & \ldots & x_N^M \\
\end{pmatrix}^T
$

```python
# 基底関数の計画行列
def phi(x, M):
    if type(x) is not np.ndarray:
        xx = np.array([[x]])
    else:
        xx = x.reshape(x.size, 1)
    poly = PolynomialFeatures(M)
    return poly.fit_transform(xx).T
```

```python
def normal(x, y, beta, SN, mN):
    M = mN.shape[0]-1 
    var = 1/beta + phi(x, M).T @ SN @ phi(x, M)
    mu = phi(x, M).T @ mN
    return np.sqrt(1/(2*np.pi*var))*np.exp(- 1 / (2*var) * ((y-mu)**2))

vnorm = np.vectorize(normal)
vnorm.excluded = {3, 4} 
```

## データ生成　（直線編）

```python
# ランダムシードを固定
np.random.seed(1234)

# y = w0 + w1*x 
M = 1

# 訓練データ数
N = 50

x = np.random.uniform(low=-0.5, high=0.5, size=N)

a_true = 0.8
b_true = 0.0
sigma = 0.2
y = a_true * x.T + b_true + sigma * np.random.randn(N)
y = y.reshape(N, 1)

Phi = phi(x, 1).T

# 事前分布の分散
alpha = 1/1000
# 想定する誤差の分散
beta = 1/0.2

# SN
SN = np.linalg.inv(alpha * np.eye(M+1) + beta * Phi.T @ Phi)

# mN
mN = beta * SN @ Phi.T @ y
```

```python
# 結果の表示

# 予測分布を求める
x_p, y_p = np.meshgrid(np.linspace(-1, 1, 100), np.linspace(1, -1, 100))

# ベクトル化した関数を呼び出す
z = vnorm(x_p, y_p, beta, SN, mN)

plt.figure(figsize=(10, 8))
plt.xlim(-1, 1)
plt.ylim(-1.0, 1.0)

xx = np.linspace(-1, 1, 128)
aa = mN[1][0]
bb = mN[0][0]
yy = aa * xx + bb
yytrue = a_true * xx + b_true

# 等高線
plt.contourf(x_p, y_p, z, 10, vmax=1.0,  alpha=0.5) #cm.gist_heatにするとかっこいい
plt.colorbar()

# MAP解
plt.plot(xx, yy, 'r-', linewidth=3)

# 真値
plt.plot(xx, yytrue, 'g-', linewidth=2)

# データ
plt.scatter(x, y)

plt.title('Bayesian Estimation')
```

## データ生成 (sin編)

```python
# x^4項まで
M = 4

# 訓練データ数
N = 20

# 訓練データの列ベクトル
x = np.random.uniform(low=-1, high=1, size=N)
# 訓練データ y の列ベクトル
y = np.sin(np.pi * x.T) + np.random.normal(0, 0.2, N)

y = y.reshape(N, 1)

# 計画行列Phiを作成
Phi = phi(x, M).T

# 事前分布の分散
alpha = 1/100000
# 想定する誤差の分散
beta = 1/0.2

# SN
SN = np.linalg.inv(alpha * np.eye(M+1) + beta * Phi.T @ Phi)

# mN
mN = beta * SN @ Phi.T @ y
```

```python
# 予測分布を求める
t = np.linspace(-1.5, 1.5, 128)
x_p, y_p = np.meshgrid(t, t)

# 確率分布を求める
z = vnorm(x_p, y_p, beta, SN, mN)

plt.figure(figsize=(10, 8))
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

xx = np.linspace(-1.5, 1.5, 128)
yy = phi(xx, M).T @ mN

ytrue = np.sin(np.pi * xx)

# 等高線
plt.contourf(x_p, y_p, z, 10, vmax=1.0,  alpha=0.5) #cm.gist_heatにするとかっこいい
plt.colorbar()

# MAP 
plt.plot(xx, yy, 'r-', linewidth=3)

# 真値
plt.plot(xx, ytrue, 'g-', linewidth=2)

# データ
plt.scatter(x, y)

plt.title('Bayesian Est. for Sine Curve')
```
