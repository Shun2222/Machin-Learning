```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns; sns.set()
```

## データ生成

```python
# N 点のデータを [0, 1) の区間に生成
# ytrue = ax + b で真値データを生成
# y = ytrue + e で観測データを生成

N = 30

a = 1.0
b = 0.0
sigma = 0.1

x = np.random.rand(N) # xに [0, 1) の乱数を N 個生成

ytrue = a * x + b #真値を生成して

y = ytrue + np.random.randn(N) * sigma #真値にノイズを載せて観測データを模擬する
```

```python
# データの表示　データを目で確認することは大切

# 観測データ(青)
plt.scatter(x, y)
plt.xlim((-0.2, 1.2))
plt.ylim((-0.2, 1.2))

# 真の直線(緑)
xx = np.linspace(-0.2, 1.2, 128)
yy = a * xx + b

plt.plot(xx, yy, 'g-')

# 座標軸の描写
plt.axhline()
plt.axvline()
```

## 前処理

```python
# 中心化 (centering)

# もとのデータを保存
xorg = x
yorg = y


# 中心座標を求める
Ex = np.mean(x)
Ey = np.mean(y)

x = x - Ex
y = y - Ey

print('中心座標は Ex={:.4f}, Ey={:.4f}'.format(Ex, Ey))
```

## 最小二乗法の準備

次に，二乗誤差を計算してみるために誤差関数 $E(a, b) = \frac{1}{2N} \sum_n (y_n - (a x_n + b))^2$ を定義して描画してみる.
$$
   E(a, b) =  \frac{1}{2N} \sum_n (y_n^2 - 2 y_n (a x_n + b) + (a^2 x_n^2 + 2 a b x_n + b^2)) \\
   = \frac{1}{2} (E_{yy} - 2 a E_{xy} - 2 b E_{y} + a^2 E_{xx} + 2 a b E_{x} + b^2) \\
   = \frac{1}{2} (E_{yy} - 2 a E_{xy} + a^2 E_{xx} + b^2)
$$

ただし，

* $E_{yy} = 1/N \sum_n y_n^2$
* $E_{xy} = 1/N \sum_n x_n y_n$
* $E_{xx} = 1/N \sum_n x_n^2$
* $E_{x} = 1/N \sum_n x_n = 0$ 中心化しているので 0 扱い
* $E_{y} = 1/N \sum_n y_n = 0$ 中心化しているので 0 扱い


```python
# 二乗誤差を求める関数
def E(a, b, xobs, yobs):
    Ex = np.mean(xobs)
    Ey = np.mean(yobs)
    Exx = np.mean(xobs**2)
    Eyy = np.mean(yobs**2)
    Exy = np.mean(xobs * yobs)
    
    return 0.5 * (Eyy - 2 * a * Exy - 2 * b * Ey + a**2 * Exx + 2 * a * b * Ex + b**2)
```

```python
# グラフを眺める（b固定）

aa = np.linspace(-4, 4, 128)

plt.xlabel('a value')
plt.ylabel('E(a, b)')
plt.title('Loss E(a, b) where b = 0 and 1')
plt.plot(aa, E(aa, b=0, xobs=x, yobs=y), 'b-')
plt.plot(aa, E(aa, b=1, xobs=x, yobs=y), 'r-')
```

```python
# グラフを眺める（a固定）
bb = np.linspace(-4, 4, 128)

plt.plot(bb, E(a = 0, b = bb, xobs=x, yobs=y))
plt.plot(bb, E(a = 1, b = bb, xobs=x, yobs=y))
plt.xlabel('b value')
plt.ylabel('E(a, b)')
plt.title('Loss E(a, b) where a = 0 and 1')
```

```python
# 等高線
t = np.linspace(-4, 4, 64)
aa, bb = np.meshgrid(t, t)

# 尤度関数
vv = np.exp( - E(aa, bb, xobs = x, yobs = y))

h = plt.contour(aa, bb, vv)

plt.xlabel('a')
plt.ylabel('b')
plt.scatter(a, b)
plt.colorbar()
plt.title('Contour of E(a, b)')
```
